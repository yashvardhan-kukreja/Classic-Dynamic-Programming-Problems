Given

total number of floors = f<br>
Numbers of eggs available = e<br>

We have to find the minimum possible number of trials to find a floor k from which egg becomes breakable.


```python
def solve(e, f):
    if f==0 or f==1:
        return f
    if e == 1:
        return f
    min_t = 999999999999
    for k in range(1, f+1):
        temp = 1 + max(solve(e-1, k-1), solve(e, f-k))
        min_t = min(min_t, temp)
    return min_t



def iterative_solve(n, k):    
    sol = [[0 for j in range(k+1)] for i in range(n+1)]
    for i in range(n+1):
        for j in range(k+1):
            if i == 1:
                sol[i][j] = j
                continue
            if j==0 or j==1:
                sol[i][j] = j
                continue
            min_t = 99999999999999
            for x in range(1, j+1):
                temp = 1 + max(sol[i-1][x-1], sol[i][j-x])
                min_t = min(min_t, temp)
            sol[i][j] = min_t
    return sol[n][k]
        
```

```python
solve(2, 36)
```
